```python
#normalizing load value
D_max_daily = data.groupby(data.index).Demand.max().to_numpy()
D_max = max(D_max_daily)
D_max
```

```python
data.Demand=data.Demand/D_max
data.head()
```

```python
df1=np.array(data["Demand"]).reshape(-1,1)
df1
```

```python
df1.shape
```

```python
##splitting dataset into train and test split
training_size=int(len(df1)*0.65)
test_size=len(df1)-training_size
train_data,test_data=df1[0:training_size,:],df1[training_size:len(df1),:1]
```

```python
training_size,test_size
```

```python
train_data
```

```python
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return np.array(dataX), np.array(dataY)
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
```

```python
print(X_train.shape), print(y_train.shape)
```

```python
print(X_test.shape), print(ytest.shape)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
X_train
```

```python
### Create the Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_absolute_percentage_error',optimizer='adam')
```

```python
model.summary()
```

```python
import time
start_time = time.time()
history=model.fit(X_train,y_train,epochs=10,batch_size=64,validation_split=0.1)
finish_time = time.time()
print("--- %s seconds ---" % (finish_time - start_time))
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend();
```

```python
### Lets Do the prediction and check performance metrics
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
train_predict=train_predict*D_max
test_predict=test_predict*D_max
```

```python
### Calculate RMSE performance metrics
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train*D_max,train_predict))
```

```python
### Test Data RMSE
math.sqrt(mean_squared_error(ytest*D_max,test_predict))
```

```python
### Plotting 
# shift train predictions for plotting
import numpy
look_back=100
trainPredictPlot = numpy.empty_like(df1)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict
# shift test predictions for plotting
testPredictPlot = numpy.empty_like(df1)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(df1)-1, :] = test_predict
# plot baseline and predictions
# plt.plot(df1*D_max)

plt.plot(trainPredictPlot,label="Training Pred")
plt.plot(testPredictPlot,label="Test Pred")
plt.legend(loc="upper left")
plt.show()
```
